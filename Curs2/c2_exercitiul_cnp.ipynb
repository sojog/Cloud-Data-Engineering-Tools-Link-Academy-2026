```python


def este_valid(cnp:str) -> bool:
  # Formatul
  if not isinstance(cnp, str): return False
  if not cnp.isdigit(): return False
  if len(cnp) != 13: return False

  gen = int(cnp[0])
  if gen not in range(1, 9): return False

  an = cnp[1:3]

  if gen in [1, 2]:
    an_cal = 1900
  elif gen in [3, 4]:
    an_cal = 1800
  if gen in [5, 6]:
    an_cal = 2000

  an_cal +=  int(an)

  # Anul trebuie sa fie cu 14 an mai mic cel putin decat anul curent
  if datetime.now().year - 14 < an_cal: return False


  luna = int(cnp[3:5])
  if luna not in range(1, 13): return False

  zi = int(cnp[5:7])
  if zi not in range(1, 32): return False

  if luna in [4,6,9,11] and zi not in range(1, 31): return False

  if luna == 2:
      if zi not in range(1, 30): return False
      este_bisect = (an_cal % 400 == 0 or (an_cal % 4 == 0 and an_cal % 100 != 0))
      if zi == 29 and not este_bisect : return False


  judet = int(cnp[7:10])
  if judet not in (list(range(1, 49)) + [51, 52, 70]): return False

  masca = "279146358279"

  suma = 0
  for i in range(12):
    suma += int(cnp[i]) * int(masca[i])

  # Rezultatele înmulțirilor sunt însumate, iar numărul rezultat se împarte la 11.
  rest = suma % 11

  print("restul :", rest)

  # Dacă restul împărțirii este mai mic decât 10, atunci acel număr reprezintă cifra de control, cea de 13-a cifra a CNP-ului.
  if rest < 10 and str(rest) != cnp[-1]: return False

  # În schimb, dacă restul împărțirii este 10, atunci cifra de control este 1.
  if rest == 10 and str(rest) != "1": return False

  return True
```

```python
def test_functie_este_valid():

  # 1. Format
  #     - sa fie cifre
  #     - 13 caractere
  assert este_valid(["1234567890123"]) == False, "Accepta doar stringuri"
  assert este_valid("qwertyuiopasd") == False, "Format este doar din cifre"
  assert este_valid("12345678901234") == False, "Fix 13 cifre"
  assert este_valid("123456789012") == False, "Fix 13 cifre"

  # 2. Cifrele
  #       - prima -> 1..8
  assert este_valid("0234567890123") == False, "Range intre 1 si 8"
  assert este_valid("9234567890123") == False, "Range intre 1 si 8"

  #      - an ->  # Anul trebuie sa fie cu 14 an mai mic cel putin decat anul curent
  assert este_valid("5164567890123") == False, "Anul trebuie sa fie cu 14 an mai mic cel putin decat anul curent"
  assert este_valid("6134567890123") == False, "Anul trebuie sa fie cu 14 an mai mic cel putin decat anul curent"

  #.   - luna intre 1 si 12
  assert este_valid("5000067890123") == False, "Range-ul lunii intre 1 si 12"
  assert este_valid("1771367890123") == False, "Range-ul lunii intre 1 si 12"
  assert este_valid("1779967890123") == False, "Range-ul lunii intre 1 si 12"
  assert este_valid("1774567890123") == False, "Range-ul lunii intre 1 si 12"

  assert este_valid("1890132890123") == False, "Luna peste 31 nu este posibil"
  assert este_valid("1890100890123") == False, "Luna peste 31 nu este posibil"
  #     - ziua intre 1 si 31 - pt 1, 3, 5, 7,8,10,12
  # assert este_valid("1890130045123") == True, "Luna peste 31 nu este posibil"
  #     - ziua intre 1 si 30 - pt 4,6,9,11
  assert este_valid("1890431890123") == False, "Lunile aprilie, iunie, septembrie si noiembrie au 30"
  #     - ziua 28 sau 29 pt 2
  assert este_valid("1890229890123") == False, "Lunile februarie are maxim 29"
  # assert este_valid("1880229008123") == True, "Lunile februarie are maxim 29"

  # 3. Judetul
  assert este_valid("1880220190123") == False, "A 8 a cifra trebuie sa fie 0"

  # 4. Validare ultima cifra
  assert este_valid("2990219045001") == False, "Ultima cifra este de control si ar trebui sa dea 0"
  assert este_valid("2990219045000") == True, "Ultima cifra este de control si ar trebui sa dea 0"


test_functie_este_valid()
```
