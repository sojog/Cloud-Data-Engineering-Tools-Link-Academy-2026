```python

def are_access_v1(varsta, gen):

  if (not isinstance(varsta, int)) or (not isinstance(gen, str)):
    return False

  gen = gen.upper()
  if gen not in ["M", "F"]:
    return False

  if varsta < 16 or varsta >= 70:
    return False

  if varsta < 18 and gen == "M":
    return False

  return True
```

```python

def are_access(varsta, gen):

  if (not isinstance(varsta, int)) or (not isinstance(gen, str)):
    return False

  gen = gen.upper()
  if gen not in ["M", "F"]:
    return False

  if varsta in range(18, 70):
    return True

  if varsta in range(16, 18) and gen == "F" :
    return True

  return False
```

```python

def test_are_access():
  assert are_access("orice", [31251] ) == False, "genul trebuie sa fie string si varsta int "
  assert are_access(1421.321, 31251 ) == False, "genul trebuie sa fie string si varsta int "

  assert are_access(32, "orice" ) == False, "genul trebuie sa fie M sau F"

  assert are_access(32, "M" ) == True, "genul trebuie sa fie M sau F"

  assert are_access(32, "m" ) == True, "genul trebuie sa fie M sau F"

  assert are_access(10, "m" ) == False, "Persoanele sub 16 nu au acces"

  assert are_access(16, "m" ) == False, "Barbatii sub 18 nu au access"


test_are_access()
```
