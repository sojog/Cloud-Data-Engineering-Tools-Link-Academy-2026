```python
def romane_to_int(text:str) -> int:
  valoare_totala = 0
  ultima_valoare = 0

  for index, ch in enumerate(reversed(text)):
    valoare = roman_map[ch]

    if ultima_valoare > valoare:
      valoare_totala -= valoare
    else:
      valoare_totala += valoare

    ultima_valoare = roman_map[ch]

  return valoare_totala
```
