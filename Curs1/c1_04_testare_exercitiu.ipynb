```python
def romane_to_int(text:str) -> int:
  roman_map = {
      'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000
  }

  for ch in text:
    if ch not in roman_map:
      return "Numar invalid"

  valoare_totala = 0
  ultima_valoare = 0

  for index, ch in enumerate(reversed(text)):
    valoare = roman_map[ch]

    if "5" in str(ultima_valoare) and "5" in str(valoare) and ultima_valoare == valoare:
      return "Numar invalid"

    if ultima_valoare > valoare:
      valoare_totala -= valoare
    else:
      valoare_totala += valoare

    ultima_valoare = roman_map[ch]

  return valoare_totala
```

```python
def romane_to_int(roman_numeral:str) -> int:
  roman_map = {
      'I': 1, 'IV':4, 'V': 5, 'IX':9, 'X': 10, 'XL':40, 'L': 50, 'XC':90, 'C': 100, 'CD':400, 'D': 500, 'CM':900, 'M': 1000
  }
  total = 0
  i = 0

  for ch in roman_numeral:
    if ch not in roman_map:
      return "Numar invalid"


  ### Nu am voie sa repet mai mult de 3 ori un caracter (I, X, C, M)
  ### Nu am voie sa dublez caracterele V, L, D

  ultima_valoare = 0

  while i < len(roman_numeral):

    valoare = roman_numeral[i]

    # Check for two-character combinations (e.g., IV, IX)
    if "5" in str(ultima_valoare) and "5" in str(valoare) and ultima_valoare == valoare:
      return "Numar invalid"

    if i + 1 < len(roman_numeral) and roman_numeral[i:i+2] in roman_map:
      valoare = roman_numeral[i:i+2]

      i += 2
    # Otherwise, process single character
    else:
      valoare = roman_numeral[i]
      i += 1
    total += roman_map[valoare]
    ultima_valoare = valoare

  return total
```
