```python
def romane_to_int(text: str) -> int:
    roman_map = {
        'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000
    }
    
    for ch in text:
        if ch not in roman_map:
            raise ValueError(f"Caracter invalid: {ch}")
            
    valoare_totala = 0
    ultima_valoare = 0
    consecutive_count = 0
    frecvente_invalide = {'V', 'L', 'D'}
    
    # 2. Parcurgere de la dreapta la stânga
    for ch in reversed(text):
        valoare = roman_map[ch]
        
        if valoare == ultima_valoare:
            consecutive_count += 1
            if consecutive_count > 3 or (ch in frecvente_invalide and consecutive_count > 1):
                raise ValueError(f"Format invalid (repetare ilegala): {text}")
        else:
            consecutive_count = 1

        if ultima_valoare > valoare:
            if valoare * 10 < ultima_valoare:
                raise ValueError(f"Scadere ilegala detectata în: {text}")
            valoare_totala -= valoare
        else:
            valoare_totala += valoare
            
        ultima_valoare = valoare
        
    # =========================================================================
    # 3. VERIFICAREA FINALĂ (Validare prin reconversie)
    # =========================================================================
    # Generăm forma romană oficială pentru 'valoare_totala' obținută mai sus
    numar = valoare_totala
    forma_oficiala = ""
    
    # Mapare descrescătoare pentru generare corectă
    valori_revers = [
        (1000, 'M'), (900, 'CM'), (500, 'D'), (400, 'CD'),
        (100, 'C'), (90, 'XC'), (50, 'L'), (40, 'XL'),
        (10, 'X'), (9, 'IX'), (5, 'V'), (4, 'IV'), (1, 'I')
    ]
    
    for val, simp in valori_revers:
        while numar >= val:
            forma_oficiala += simp
            numar -= val

    # Dacă string-ul inițial nu este identic cu forma lui oficială, înseamnă că structura a fost greșită
    if text != forma_oficiala:
        raise ValueError(f"Structură sintactică romană ilegală: '{text}' (Ar fi trebuit să fie '{forma_oficiala}')")
    
    return valoare_totala


# --- SCRIPT DE TESTARE ---
teste_bune = ["III", "IV", "XIX", "MMXXVI", "MCMXCIV"]
teste_rele = ["IIV", "IVX", "MCMM", "VV", "MMCMD"]

print("--- Teste valide (Așteptat: Succes) ---")
for t in teste_bune:
    print(f"{t} -> {romane_to_int(t)}")

print("\n--- Teste invalide (Așteptat: Prindere Eroare) ---")
for t in teste_rele:
    try:
        rezultat = romane_to_int(t)
        print(f" PASSED (Greșit!) -> {t} a returnat {rezultat}")
    except ValueError as e:
        print(f" BLOCAT CORECT -> {t} | Motiv: {e}")
```
