# 1. citire din S3 bucket

# pip install boto

```python
import boto3
from botocore.config import Config
from botocore import UNSIGNED
```

```python
s3 = boto3.client("s3", config=Config(signature_version=UNSIGNED))

BUCKET_NAME = "noaa-global-hourly-pds"

file_dict = s3.list_objects_v2(Bucket=BUCKET_NAME)['Contents']
file_dict
```

```python
file_names = []
for d in file_dict:
    file_name = d["Key"]
    file_names.append(file_name)
```

```python
first_name:str = sorted(file_names)[0]
YEAR_TO_INSERT = first_name.split("/")[0]
YEAR_TO_INSERT
```

# 2. Verficare pe site

```python
URL = "https://www.bereciucas.ro/"
```

```python
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
```

```python
browser = Chrome()
browser.get(URL)
```

```python
COOK_POL_ID = "CybotCookiebotDialogBodyLevelButtonLevelOptinAllowAll"
```

```python
browser.find_element(By.ID, COOK_POL_ID).click()
```

### Inserarea intr-o celula anume

```python
browser.find_element(By.NAME, "year_1").send_keys("1")
```

```python
for index, ch in enumerate(YEAR_TO_INSERT):
    print("index:", index, "ch", ch)
```

```python
for index, ch in enumerate(YEAR_TO_INSERT, start=1):
    browser.find_element(By.NAME, f"year_{index}").send_keys(ch)
```
