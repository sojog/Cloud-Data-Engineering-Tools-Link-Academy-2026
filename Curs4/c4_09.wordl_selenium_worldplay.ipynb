```python
LITERE = ['qwertyuiop','asdfghjkl','-zxcvbnm']


def get_row_col_for(chr):
    for index_row, row in enumerate(LITERE):
        if chr in row:
            index_col = row.index(chr)
            return index_row + 1, index_col + 1
```

```python
print(get_row_col_for("d"))
```

```python
def generate_xpath(chr):
    row, column = get_row_col_for(chr)
    return f'//*[@id="__next"]/div[1]/div[1]/div/div[2]/div/div/div[{row}]/button[{column}]'
generate_xpath("d")
```

```python
import time
```

```python
generate_xpath("a")
```

```python
for word in ["audio", "stern", "glyph"]:
    for ch in word + "-":
        browser.find_element(By.XPATH, generate_xpath(ch)).click()
        time.sleep(1)
```
