```python
from selenium import webdriver
from selenium.webdriver.common import by
```

```python
URL = "https://link-academy.com"
```

```python
from selenium.webdriver import ChromeOptions
```

```python
options  = ChromeOptions()
options.add_argument("--incognito")
options.add_argument("--window-size=1920,1080")
options.add_argument("--headless=true")
```

```python
browser = webdriver.Chrome(options=options)
browser.get(URL)
```

```python
TEL_XPATH = '//*[@id="mainSidebar"]/div[1]/div[2]/p[2]'
```

```python
tel_element = browser.find_element(by.By.XPATH, TEL_XPATH)
tel_element.text
```

```python
with open("numere_de_telefon.txt", "w") as file_writer:
    file_writer.write(tel_element.text)
```

```python
browser.quit()
```
