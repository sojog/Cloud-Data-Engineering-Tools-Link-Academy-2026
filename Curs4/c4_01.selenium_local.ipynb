```python
from selenium import webdriver
```

```python
from selenium.webdriver.common import by
```

```python
ABOUT_XPATH =   '//*[@id="navLeft"]/ul/li[6]/a'
```

```python
browser = webdriver.Chrome()
browser.get("https://link-academy.com")
```

```python
about_link = browser.find_element(by.By.XPATH, ABOUT_XPATH)
about_link.click()
```

```python
TEL_XPATH = '//*[@id="mainSidebar"]/div[1]/div[2]/p[2]'
```

```python
tel_element = browser.find_element(by.By.XPATH, TEL_XPATH)
tel_element.text
```

```python
with open("numere_de_telefon.txt", "w") as file_writer:
    file_writer.write(tel_element.text)
```

```python
browser.quit()
```
