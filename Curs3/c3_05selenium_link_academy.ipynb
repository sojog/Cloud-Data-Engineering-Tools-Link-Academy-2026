```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
browser = webdriver.Chrome()
```

```python

URL = "https://www.link-academy.com/cursuri-de-specializare-in-domeniul-it-6-departamante"
browser.get(URL)
```

```python

cookie_accept_button = browser.find_element(By.ID, "politikaOK")
cookie_accept_button.click()
```

```python

close_button = browser.find_element(By.CLASS_NAME, "zatvori")
close_button.click()
```

```python


program_link = browser.find_element(By.LINK_TEXT, "AI & Python Program")
program_link.click()
```

```python

python_course_link = browser.find_element(By.XPATH, "/html/body/div[2]/div[1]/div[2]/table[1]/tbody/tr[2]/td[3]/a")
python_course_link.click()

```

```python
browser.quit()
```
