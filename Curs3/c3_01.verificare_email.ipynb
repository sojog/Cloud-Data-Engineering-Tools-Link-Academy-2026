```python
def este_valid(email):
  # Lungimea maximă: 255 de caractere (inclusiv local-part și @)
  # @ (simbolul "at"): Trebuie să apară exact o dată
  if not isinstance(email, str) or not email or len(email) > 255 or email.count("@") != 1 :
    return False

  username, domain = email.split("@")
  # Lungimea maximă: 64 de caractere
  # Nu poate începe sau se termina cu un punct (.)
  # Nu poate avea două puncte consecutive (..)
  if not username or len(username) > 64 or username[0] == "."  or username[-1] == "."  or ".." in username:
    return False

  # Poate conține litere (a-z, A-Z), cifre (0-9) și următoarele caractere speciale: . +
  if not all(ch in string.ascii_letters + string.digits + ".+" for ch in username):
    return False



  # domain (domeniul):

   # Trebuie să aibă cel puțin un punct (.) (de exemplu: example.com)
    # Nu poate începe sau se termina cu un punct (.) sau cratimă (-)
  if not domain.count(".") or  not domain or domain[0] == "." or domain[-1] == "." :
    return False

  # Poate conține litere (a-z, A-Z), cifre (0-9), puncte (.) și cratime (-)
  if not all(ch in string.ascii_letters + string.digits + ".-" for ch in domain):
      return False

  name, extention = domain.split(".")

  if not name or not extention or len(extention) < 2:
    return False

  return True

```

```python
def test_validare_email():

  assert este_valid(421) == False

  assert este_valid("") == False

  assert este_valid("hello" * 255 + "@test.com") == False

  assert este_valid("hello" + "test.com") == False

  assert este_valid("@test.com") == False

  assert este_valid("a" * 65 + "@test.com") == False

  assert este_valid(".hello@test.com") == False

  assert este_valid("hello.@test.com") == False
  assert este_valid("hel..lo@test.com") == False

  assert este_valid("hel%.lo@test.com") == False
  assert este_valid("hel .lo@test.com") == False


  assert este_valid("hel%.lo@test.com") == False
  assert este_valid("hello@.test.com") == False
  assert este_valid("hello@.test.com.") == False
  assert este_valid("hello@testcom") == False
  assert este_valid("hello@testcom.p") == False

test_validare_email()
```
