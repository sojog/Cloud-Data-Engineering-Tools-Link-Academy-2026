```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
browser = webdriver.Chrome()
```

```python
URL = "https://olx.ro"
# URL = "https://emag.ro"
```

```python
browser.get(URL)
```

```python
accept_button = browser.find_element(By.ID, "onetrust-accept-btn-handler")
accept_button
```

```python
accept_button.click()
```

```python
search_input = browser.find_element(By.ID, "search")
search_input
```

```python
search_input.send_keys("masini")
search_input.send_keys(Keys.ENTER)
```

```python
rezultate = browser.find_elements(By.CLASS_NAME, "css-1tqlkj0")
rezultate 
```

```python
rezultate[0].get_attribute("href")
```

```python
for i in rezultate:
    print(i.get_attribute("href"))
```

```python
browser.quit()
```
