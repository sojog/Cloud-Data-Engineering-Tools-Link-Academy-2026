```python
import pandas as pd
```

```python
names = []
addresses = []
emails = []
```

```python
for i in range(10):
    names.append(faker_object.name())
    addresses.append(faker_object.address())
    emails.append(faker_object.email())
```

```python
df = pd.DataFrame({
    "names":names, "addresses": addresses, "emails": emails
})
df
```

```python
df.to_csv("people.csv")
```
