```python
def convert(roman_numeral):
  roman_map = {
      'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000,
      'IV': 4, 'IX': 9, 'XL': 40, 'XC': 90, 'CD': 400, 'CM': 900
  }
  total = 0
  i = 0
  while i < len(roman_numeral):
    # Check for two-character combinations (e.g., IV, IX)
    if i + 1 < len(roman_numeral) and roman_numeral[i:i+2] in roman_map:
      total += roman_map[roman_numeral[i:i+2]]
      i += 2
    # Otherwise, process single character
    else:
      total += roman_map[roman_numeral[i]]
      i += 1
  return total

print(convert("III"))
print(convert("LVIII"))
print(convert("MCMXCIV"))
```
