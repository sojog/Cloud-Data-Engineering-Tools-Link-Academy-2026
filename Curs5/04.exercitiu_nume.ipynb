# Partea 2 - accesarea unui API folosind requests

```python
import requests
import time
```

```python
response = requests.get(f"https://api.nationalize.io/?name={today_names[0]}")
response
```

```python
for item in response.json()['country']:
    print(item["country_id"])
```

```python
name_dict = {}
for name in today_names:
    response = requests.get(f"https://api.nationalize.io/?name={name}")
    
    countries = []
    
    for item in response.json()['country']:
        countries.append(item["country_id"])

    name_dict[name] = countries
    time.sleep(1)

name_dict
```

# Partea 3 - accesarea unui S3 bucket folosind pandas

```python
BUCKET_NAME = "noaa-ghcn-pds"
```

```python
import boto3
from botocore.config import Config
from botocore import UNSIGNED
```

```python
s3 = boto3.client("s3", config=Config(signature_version=UNSIGNED))
BUCKET_NAME = "noaa-ghcn-pds"

objects_list = s3.list_objects_v2(Bucket=BUCKET_NAME)['Contents']

for d in objects_list:
    if d["Key"].endswith(".text"):
        print(d["Key"])
```

```python
import pandas as pd
```

```python
pd.read_csv('https://noaa-ghcn-pds.s3.amazonaws.com/csv/by_year/1763.csv')
```

```python
BUCKET_NAME = "noaa-ghcn-pds"
FILE_NAME = "ghcnd-countries.txt"

returned_object = s3.get_object(Bucket=BUCKET_NAME, Key=FILE_NAME)
returned_object['Body']
```

```python
response_bytes =  returned_object['Body'].read()
response_bytes
```

```python
type(response_bytes.decode("utf-8"))
```

```python
file_content = response_bytes.decode("utf-8")
file_content
```

```python
file_lines = file_content.split("\n")
file_lines
```

```python
country_dict = {}

for line in file_lines:
    try:
        country_code, country_name = line.split(" ", 1)
        country_dict[country_code] = country_name
    except:
        print("Linia incorecta", line)
```

```python
country_dict
```

```python
name_dict
```

```python
full_name_dict = {}

for name in name_dict:
    full_name_dict[name] = []
    for ccode in name_dict[name]:
        cname = country_dict.get(ccode)
        if cname:
            full_name_dict[name].append(cname.strip())
        else:
            full_name_dict[name].append(ccode)
full_name_dict          
```
