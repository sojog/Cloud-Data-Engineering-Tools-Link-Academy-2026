```python
import boto3
from botocore.config import Config
from botocore import UNSIGNED
```

```python
s3 = boto3.client("s3", config=Config(signature_version=UNSIGNED))
s3
```

```python
BUCKET_NAME = "noaa-ghcn-pds"

s3.list_objects_v2(Bucket=BUCKET_NAME)

```

```python
objects_response = s3.list_objects_v2(Bucket=BUCKET_NAME)


file_list = []

if objects_response.get('Contents'):
    for obj in objects_response['Contents']:
        file_list.append(obj['Key'])
else:
    print("Bucket-ul este gol sau nu exista acces la el")

len(file_list)
```

```python
BUCKET_NAME = "noaa-ghcn-pds"
FILE_NAME = 'csv.gz/by_station/ASN00005011.csv.gz'

returned_object = s3.get_object(Bucket=BUCKET_NAME, Key=FILE_NAME)
returned_object['Body']
```

```python
response_bytes =  returned_object['Body'].read()
response_bytes
```

```python
type(response_bytes)
```

```python
import gzip

raw_value = gzip.decompress(response_bytes)
raw_value
```

```python
import csv

content = raw_value.decode("utf-8")
type(content)
```

```python
import io
```

```python
reader = csv.DictReader(io.StringIO(content))
reader
```

```python
rows = list(reader)
rows
```

```python
import pandas as pd
pd.DataFrame(rows)
```
